## Цель и задачи исследования

**Заказчик исследования:** каршеринговая компания.

**Цель:** создать систему, которая могла бы оценить риск ДТП по выбранному маршруту движения.

**Задачи (этапы) исследования:**
1. Провести загрузку и общий обзор данных:
- загрузить таблицы SQL;
- провести первичное исследование таблиц
2. Провести статистический анализ факторов ДТП:
- определить месяцы, в которые происходит наибольшее количество аварий;
- создать аналитические задачи для коллег (не менее 6);
- прописать порядок решения задач (не менее 2).
3. Провести предобработку и исследовательский анализ данных:
- выявить и устранить пропуски;
- выявить и устранить дубликаты в данных;
- изменить типы данных при необходимости.
- провести статистический анализ признаков;
- изучить параметры распределения признаков (графический анализ);
- провести корреляционный анализ.
4. Провести подготовку данных для обучения модели:
- закодировать категориальные признаки;
- провести масштабирование числовых признаков.
5. Построить модели прогнозирования риска ДТП:
- построить не менее 3 моделей;
- определить метрику для оценки моделей;
- осуществить перебор параметров;
- провести обучение нейросети;
- построить график «Факт — Прогноз».
6. Оценить лучшую модель:
- провести графический анализ «Матрица ошибок»;
- проанализировать важность основных факторов, влияющих на вероятность ДТП;
- для одного из выявленных важных факторов провести дополнительное исследование.
7. Сформулировать общие выводы.

### Решения задач

**Задача 1.**

Проанализируйте взаимосвязь категории нарушения и возраста автомобиля.

Алгоритм:
- с помощью SQL-запроса объедините таблицы collisions и vehicles;
- сгруппируйте данные по возрасту автомобиля и категориям нарушений и посчитайте число ДТП;
- постройте тепловую карту;
- сделайте выводы.

```python
query = '''
    SELECT vehicle_age, pcf_violation_category, COUNT (DISTINCT case_id)
    FROM vehicles
    LEFT JOIN collisions USING(case_id)
    WHERE vehicle_age IS NOT NULL
    GROUP BY vehicle_age, pcf_violation_category
    ORDER BY vehicle_age, pcf_violation_category;
'''

df = pd.read_sql_query(query, con = engine)
```

```python
df['vehicle_age'] = pd.cut(
    df['vehicle_age'],
    bins=[-1, 4, 10, 15, 20, 200],
    labels=['0-4', '5-9', '10-14', '15-19', '20+']
) # сгруппируем возраст в интервалы
df = df.pivot_table(
    values='count',
    index='vehicle_age',
    columns='pcf_violation_category',
    aggfunc='sum',
    fill_value=0,
    margins=True
)

df = df.div(df.iloc[:,-1]/100, axis=0).round(2).drop('All').drop('All', axis=1)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df, annot=True, fmt='.1f', cmap='coolwarm');
fig, ax = plt.gcf(), plt.gca()
fig.suptitle('Доля нарушений по типам в зависимости от возраста автомобиля', va='bottom', y=0.9, fontsize=16)
ax.set_xlabel('Тип нарушения')
ax.set_ylabel('Возраст автомобиля')
cb_ax = fig.axes[1]
cb_ax.set_ylabel('Доля нарушений, %')
plt.show()
```

Самый частый тип нарушения - превышение скорости. При этом он чаще встречается, если автомобиль возрастом 5-9 лет.

**Задача 2.**

Проанализируйте взаимосвязь размера ущерба от ДТП и трезвости водителя.

Алгоритм:
- с помощью SQL-запроса объедините таблицы collisions и parties;
- сгруппируйте данные по трезвости водителя и размеру ущерба и посчитайте число ДТП;
- постройте график;
- сделайте выводы.

```python
query = '''
    SELECT party_sobriety, collision_damage, COUNT (DISTINCT case_id)
    FROM parties
    LEFT JOIN collisions USING(case_id)
    GROUP BY party_sobriety, collision_damage
    ORDER BY party_sobriety, collision_damage;
'''

df = pd.read_sql_query(query, con = engine)
```

```python
df = df.fillna('None')
df['share'] = df['count'] / df.groupby('party_sobriety')['count'].transform(sum)
```

```python
sns.barplot(
    df,
    y='party_sobriety',
    x='share',
    hue='collision_damage',
    hue_order=['scratch', 'small damage', 'middle_damage', 'severe damage', 'fatal']
)
plt.title('Зависимость размера ущерба от трезвости водителя')
plt.xlabel('Доля ДТП')
plt.ylabel('Трезвость водителя')
plt.legend(loc='upper left',
                   bbox_to_anchor=(1, 1))
plt.show()
```

На графике видно, что доля ДТП с серьезными и фатальными повреждениями выше у нетрезвых водителей.

**Задача 4 (дополнительно).**

Определите, автомобили какого возраста чаще становятся виновниками ДТП.

Алгоритм:
- с помощью SQL-запроса объедините таблицы parties и vehicles;
- сгруппируйте данные по возрасту автомобиля и посчитайте долю ДТП, где автомобиль является виновником;
- постройте график;
- сделайте выводы.

```python
query = '''
    SELECT vehicle_age, at_fault, COUNT (DISTINCT case_id)
    FROM vehicles
    LEFT JOIN parties USING(case_id, party_number)
    WHERE vehicle_age IS NOT NULL
    GROUP BY vehicle_age, at_fault;
'''

df = pd.read_sql_query(query, con = engine)
```

```python
df['fault_share'] = df['count'] / df.groupby('vehicle_age')['count'].transform(sum) * 100
df = df[df['at_fault']==1]
```

**Задача 6 (дополнительно).**

Проанализируйте динамику ДТП по годам и месяцам.

Алгоритм:
- с помощью SQL-запроса в таблице collisions сгруппируйте данные по усеченной дате и посчитайте число ДТП;
- постройте график;

###  Описательная статистика и графики параметров

Определим функции для построения графиков признаков.

```python
# Функция для построения графиков числовых непрерывных параметров.
# Функция строит "сетку" графиков для различных столбцов датафрейма.
# Параметры функции:
# df - датафрейм с исходными данными;
# columns - список столбцов, по которым строятся графики;
# num_cols - число столбцов "сетке" графиков;
# title - общий заголовок для "сетки";
# hue - столбец, для уникальных значений которого строятся дополнительные графики

def plot_quant_interval(df, columns, num_cols, title, hue=None):
    num_rows = -(-len(columns) // num_cols) # округление вверх

    fig = plt.figure(figsize=(num_cols*5, num_rows*3))
    outer_grid = fig.add_gridspec(num_rows, num_cols, hspace=0.5, wspace=0.2)

    i = 0

    color = '0.8' if hue else 'C0'

    for column in columns:
        inner_grid = outer_grid[i].subgridspec(2, 1, height_ratios=[2, 1],
                                               hspace=0, wspace=0)
        ax = inner_grid.subplots(sharex=True)
        sns.histplot(data=df, x=column, stat='percent', common_norm=False, bins=20,
                     color=color, legend=False, ax=ax[0], label='общая совокупность')\
                    .set(ylabel='Частотность/плотность\n (в рамках группы), %')

        ax[0].set_title(column, fontsize=12)
        ax2 = ax[0].twinx()

        if hue:
            sns.kdeplot(data=df, x=column, common_norm=False, hue=hue, fill=True, cut=0,
                        legend=(not i), ax=ax2)

        sns.boxplot(data=df, x=column, legend=False, ax=ax[1]).set(xlabel='Значения')
        ax2.set(ylabel='', yticks=[])

        # получаем параметры легенды первого графика и убираем ее
        if hue and not i:
            lines, labels = ax[0].get_legend_handles_labels()
            legend = ax2.get_legend()
            handles = lines + legend.legend_handles
            texts = labels + [hue + ': ' + x.get_text() for x in legend.get_texts()]
            legend.remove()

        i += 1

    # размещаем общую легенду для сетки
    if hue:
        fig.legend(handles, texts, loc='upper center',
                   bbox_to_anchor=(0.5, -0.2), fontsize=12)

    fig.suptitle(title, va='bottom', size=16, y=1.1)
    plt.subplots_adjust(
        left=0,
        right=1,
        top=1,
        bottom=0,
        wspace=0,
        hspace=0
    )
    plt.show()
```

```python
# Функция для построения графиков качественных параметров.

def plot_qual(df, columns, num_cols, title, hue=None, top=None):
    num_rows = -(-len(columns) // num_cols) # округление вверх
    fig = plt.figure(figsize=(num_cols*8, num_rows*5), layout='constrained')

    i = 0

    color = '0.8' if hue else 'C0'

    for column in columns:
        ax = fig.add_subplot(num_rows, num_cols, i+1)
        top_names = df[column].value_counts().iloc[:top].index
        data = df[df[column].isin(top_names)]

        sns.countplot(data=data, y=column, stat='percent', legend=False, ax=ax,
                      color=color,  edgecolor='black', order=top_names,
                      label='общая совокупность') \
                     .set(xlabel='Частотность, %', ylabel='Значения', title=column);

        ax.set_title('Признак: '+column, fontsize=12)

        if hue:
            sns.histplot(data=data, y=column, stat='percent', legend=(not i),
                         common_norm=False, ax=ax, hue=hue, multiple='dodge',
                         shrink=.5)

        if hue and not i:
            lines, labels = ax.get_legend_handles_labels()
            legend = ax.get_legend()
            handles = lines + legend.legend_handles
            texts = labels + [hue + ': ' + x.get_text() for x in legend.get_texts()]
            legend.remove()

        i += 1

    if hue:
        fig.legend(handles, texts, loc='upper center',
                   bbox_to_anchor=(0.5, 0), fontsize=12)

    fig.set_constrained_layout_pads(wspace=0.1, hspace=0.1)
    fig.suptitle(title, va='bottom', size=16)
    plt.show()
```

Сформируем архитектуру нейросети: 1 скрытый слой из 32 нейронов.

```python
class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size_1, output_size):
        super(NeuralNet, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size_1)
        self.act1 = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size_1, output_size)
        self.act2 = nn.Sigmoid()

    def forward(self, x):
        out = self.fc1(x)
        out = self.act1(out)
        out = self.fc2(out)
        out = self.act2(out)
        return out
```

```python
# Функция взята отсюда: https://gist.github.com/peterdhansen/ca87cc1bfbc4c092f0872a3bfe3204b2
def combine_one_hot(shap_values, name, mask, return_original=True):
    """  shap_values: an Explanation object
          name: name of new feature
          mask: bool array same lenght as features

         This function assumes that shap_values[:, mask] make up a one-hot-encoded feature
    """
    mask = np.array(mask)
    mask_col_names = np.array(shap_values.feature_names, dtype='object')[mask]

    sv_name = shap.Explanation(shap_values.values[:, mask],
                               feature_names=list(mask_col_names),
                               data=shap_values.data[:, mask],
                               base_values=shap_values.base_values,
                               display_data=shap_values.display_data,
                               instance_names=shap_values.instance_names,
                               output_names=shap_values.output_names,
                               output_indexes=shap_values.output_indexes,
                               lower_bounds=shap_values.lower_bounds,
                               upper_bounds=shap_values.upper_bounds,
                               main_effects=shap_values.main_effects,
                               hierarchical_values=shap_values.hierarchical_values,
                               clustering=shap_values.clustering,
                               )

    new_data = (sv_name.data * np.arange(sum(mask))).sum(axis=1).astype(int)

    svdata = np.concatenate([
        shap_values.data[:, ~mask],
        new_data.reshape(-1, 1)
    ], axis=1)

    if shap_values.display_data is None:
        svdd = shap_values.data[:, ~mask]
    else:
        svdd = shap_values.display_data[:, ~mask]

    svdisplay_data = np.concatenate([
        svdd,
        mask_col_names[new_data].reshape(-1, 1)
    ], axis=1)

    new_values = sv_name.values.sum(axis=1)
    svvalues = np.concatenate([
        shap_values.values[:, ~mask],
        new_values.reshape(-1, 1)
    ], axis=1)
    svfeature_names = list(np.array(shap_values.feature_names)[~mask]) + [name]

    sv = shap.Explanation(svvalues,
                          base_values=shap_values.base_values,
                          data=svdata,
                          display_data=svdisplay_data,
                          instance_names=shap_values.instance_names,
                          feature_names=svfeature_names,
                          output_names=shap_values.output_names,
                          output_indexes=shap_values.output_indexes,
                          lower_bounds=shap_values.lower_bounds,
                          upper_bounds=shap_values.upper_bounds,
                          main_effects=shap_values.main_effects,
                          hierarchical_values=shap_values.hierarchical_values,
                          clustering=shap_values.clustering,
                          )
    if return_original:
        return sv, sv_name
    else:
        return sv

shap_values_rev = shap_values
for column in cat_columns:
    shap_values_rev, sv_ms = combine_one_hot(shap_values_rev, column, [column in n for n in shap_values_rev.feature_names])

shap.plots.beeswarm(shap_values_rev, max_display=20)
```

```python
query = '''
    SELECT vehicle_age, at_fault, COUNT (DISTINCT case_id)
    FROM parties
    INNER JOIN vehicles USING(case_id, party_number)
    LEFT JOIN collisions USING(case_id)
    WHERE EXTRACT(YEAR FROM collision_date) = 2012
      AND collision_damage != 'scratch'
      AND party_type = 'car'
      AND case_id IN
      (
       SELECT DISTINCT case_id
       FROM parties
       WHERE party_type = 'car' AND at_fault = 1
      )
    GROUP BY vehicle_age, at_fault;
'''

df = pd.read_sql_query(query, con = engine)
df['fault_share'] = df['count'] / df.groupby('vehicle_age')['count'].transform(sum) * 100
df = df[df['at_fault']==1]
df = df.dropna()
df['vehicle_age'] = df['vehicle_age'].astype(int)

sns.barplot(
    df,
    x='vehicle_age',
    y='fault_share'
)
plt.title('Доля "виновных" ДТП в зависимости от возраста автомобиля')
plt.xlabel('Возраст автомобиля')
plt.ylabel('Доля "виновных" ДТП, %')
plt.show()
```

Мы видим, что вероятность стать виновником ДТП растет к 3 годам возраста автомобиля, затем снижается и снова растет к 15 годам. Видимо, автомобили "среднего возраста" уже проверены и не подбрасывают "неожиданностей" на дороге, но еще не настолько стары, чтобы выходили из строя узлы и агрегаты.
