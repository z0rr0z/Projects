Также в тестовой выборке видны различные типы пропусков в столбцах 'dept', 'level' и 'workload': nan и пробел. Учтем это при формировании пайплайна.

Разделим признаки на 3 группы: качественные (ohe_columns), качественные порядковые (ord_columns) и количественные (num_columns). Необоходимо отметить, что employment_years и supervisor_evaluation по сути тоже являются качественными порядковыми признаками, однако для целей моделирования их можно рассматривать как количественные, поскольку при их кодировании мы получим такой же результат, как и без кодирования (возможно, только с определенным линейным сдвигом). Более того, в случае пропуска какого-либо значения среди уникальных значений данных признаков (например, после 4 сразу 6) кодировщик нивелирует этот скачок (т.е. закодирует 4, например, как 4, а 6 - как 5), что может привести к потери важной для модели информации. Эти признаки отдельно поместим в num_columns_discrete, а "salary" - в num_columns_interval

```python
ohe_columns = ['dept', 'last_year_promo', 'last_year_violations']
ord_columns = ['level', 'workload']
num_columns_discrete = ['employment_years', 'supervisor_evaluation']
num_columns_interval = ['salary']
num_columns = num_columns_discrete + num_columns_interval
```

Создадим пайплайн для замены пропусков. Будем заменять их (пропуски) на наиболее часто встречающееся значение, т.к. пропусков относительно немного и такая замена сильно не повлияет на распределение значений признаков. Учитывая, что у нас два типа пропусков, в пайплайне замену необходимо провести два раза.

```python
imputer_pipe = Pipeline(
    [('simpleImputer1', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('simpleImputer2', SimpleImputer(missing_values=' ', strategy='most_frequent'))
    ]
)
```

#### Описательная статистика и графики числовых параметров

Определим функцию для построения графиков числовых непрерывных параметров.

```python
# Функция строит "сетку" графиков для различных столбцов датафрейма.
# Параметры функции:
# df - датафрейм с исходными данными;
# columns - список столбцов, по которым строятся графики;
# num_cols - число столбцов "сетке" графиков;
# title - общий заголовок для "сетки";
# hue - столбец, для уникальных значений которого строятся дополнительные графики

def plot_quant_interval(df, columns, num_cols, title, hue=None):
    num_rows = -(-len(columns) // num_cols) # округление вверх

    fig = plt.figure(figsize=(num_cols*5, num_rows*3))
    outer_grid = fig.add_gridspec(num_rows, num_cols, hspace=0.5, wspace=0.2)

    i = 0

    color = '0.8' if hue else 'C0'

    for column in columns:
        inner_grid = outer_grid[i].subgridspec(2, 1, height_ratios=[2, 1],
                                               hspace=0, wspace=0)
        ax = inner_grid.subplots(sharex=True)
        sns.histplot(data=df, x=column, stat='percent', common_norm=False, bins=20,
                     color=color, legend=False, ax=ax[0], label='общая совокупность')\
                    .set(ylabel='Частотность/плотность\n (в рамках группы), %')

        ax[0].set_title('Признак: '+column, fontsize=12)
        ax2 = ax[0].twinx()

        if hue:
            sns.kdeplot(data=df, x=column, common_norm=False, hue=hue, fill=True, cut=0,
                        legend=(not i), ax=ax2)

        sns.boxplot(data=df, x=column, legend=False, ax=ax[1])\
                    .set(xlabel='Значения')
        ax2.set(ylabel='', yticks=[])

        # получаем параметры легенды первого графика и убираем ее
        if hue and not i:
            lines, labels = ax[0].get_legend_handles_labels()
            legend = ax2.get_legend()
            handles = lines + legend.legend_handles
            texts = labels + [hue + ': ' + x.get_text() for x in legend.get_texts()]
            legend.remove()

        i += 1

    # размещаем общую легенду для сетки
    if hue:
        fig.legend(handles, texts, loc='upper center',
                   bbox_to_anchor=(0.5, -0.2), fontsize=12)

    fig.suptitle(title, va='bottom', size=16, y=1.1)
    plt.subplots_adjust(
        left=0,
        right=1,
        top=1,
        bottom=0,
        wspace=0,
        hspace=0
    )
    plt.show()
```

Выведем описательные статистики и построим графики для числовых параметров в тренировочном датасете, в графиках также рассмотрим различия распределений параметров зависимости от целевого признака (удовлетворенность сотрудника).

###  Подготовка данных

Построим пайплайны для подготовки данных исходя из категории признака:
- для качественных - замена пропусков (ранее определен пайплайн imputer) и кодировка с помощью OneHotEncoder;
- для качественных порядковых - замена пропусков и кодировка с помощью OrdinalEncoder (необходимо передать списки для правильного кодирования);
- для количественных - MinMaxScaler (чтобы "уложить" все признаки в шкалу от 0 до 1).

```python
ohe_pipe = Pipeline(
    [
        ('imputer', imputer_pipe),
        ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
    ]
)
```

```python
ord_pipe = Pipeline(
    [
        ('imputer', imputer_pipe),
        ('ord', OrdinalEncoder
         (
             categories=[
                 ['junior', 'middle', 'senior'],
                 ['low', 'medium', 'high']
             ],
             handle_unknown='use_encoded_value', unknown_value=np.nan
         )
        )
    ]
)
```

```python
num_pipe = Pipeline(
    [('scaler', MinMaxScaler())
    ]
)
```

Соберем общий пайплайн подготовки данных с помощью ColumnTransformer.

```python
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', num_pipe, num_columns)
    ],
    remainder='passthrough'
)
```

### Обучение моделей

В соотвествии с поставленной задачей определим целевую метрику - SMAPE.

```python
def SMAPE_score(y, y_pred):
    y = list(y)
    n = len(y)
    sum = 0
    for i in range(n):
        sum += abs(y[i] - y_pred[i])  / ((abs(y[i]) + abs(y_pred[i])) / 2)

    return 100 / n * sum
```

Создадим "оценщик" с учетом того, что лучшим считается меньшее значение метрики.

```python
SMAPE = make_scorer(SMAPE_score, greater_is_better=False)
```

Создадим словарь для подбора параметров:
- модели: DecisionTreeRegressor и LinearRegression;
- гиперпараметры для DecisionTreeRegressor: макс. глубина - от 5 до 10, мин. число объектов в узле - от 2 до 4, мин. число объектов в листе - от 1 до 4;
- для линейной регрессии обойдемся без подбора параметров.

```python
param_grid = [
    # словарь для модели DecisionTreeRegressor()
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(5, 11),
        'models__min_samples_split': range(2, 5),
        'models__min_samples_leaf': range(1, 5)
    },

    # словарь для модели LinearRegression()
    {
        'models': [LinearRegression()]
    },
]
```

Сформируем общий пайплайн.

```python
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeRegressor(random_state=RANDOM_STATE))
])
```

Выделим в тренировочных данных факторные и целевой признаки.

```python
X_train = train_job_satisfaction_rate.drop('job_satisfaction_rate', axis=1)
y_train = train_job_satisfaction_rate['job_satisfaction_rate']
```

Запустим сплошной подбор параметров с помощью GridSearchCV с метрикой SMAPE.

```python
grid_search = GridSearchCV(
    pipe_final,
    param_grid,
    cv=5,
    n_jobs=-1,
    scoring=SMAPE,
)
```

```python
%%time
grid_search.fit(X_train, y_train)
```

Лучшая модель - регрессия на основе дерева решений. Такой результат достаточно логичен: в ходе исследовательского анализа мы установили, что явные линейные зависимости не прослеживаются, поэтому линейная регрессия здесь показала не лучшие результаты.

Чтобы не быть голословным, проверим этот вывод с помощью статистического теста.

Проверим гипотезу о равенстве средней удовлетворенности работой у уволившихся (satisfaction_mu_quit) и оставшихся сотрудников (satisfaction_mu_stay).

Нулевая гипотеза: средняя удовлетворенность работой у уволившихся сотрудников равна средней удовлетворенности работой у оставшихся сотрудников.

Альтернативная гипотеза: средняя удовлетворенность работой у уволившихся сотрудников меньше средней удовлетворенности работой у оставшихся сотрудников.

H0: satisfaction_mu_quit = satisfaction_mu_stay

H1: satisfaction_mu_quit < satisfaction_mu_stay

Разделим датесет на два: уволившиеся (quit) и оставшиеся (stay) сотрудники.

```python
quit = test_target_quit_satisfaction_rate[test_target_quit_satisfaction_rate['quit']=='yes']['job_satisfaction_rate']
stay = test_target_quit_satisfaction_rate[test_target_quit_satisfaction_rate['quit']=='no']['job_satisfaction_rate']
```

Поскольку в данном случае мы имеем две независимые выборки, необходимо использовать t-тест о равенстве средних генеральных совокупностей.

Проверим дисперсии выборок.

```python
print(f'Дисперсия уровня удовлетворенности по уволившимся сотрудникам: {quit.var():4.3f}')
print(f'Дисперсия уровня удовлетворенности по оставшимся сотрудникам: {stay.var():6.3f}')
```

Значимых отличий дисперсии не видно, поэтому в t-тесте будем предполагать их равенство.

```python
alpha = 0.01
results = st.ttest_ind(quit, stay,
                       alternative='less')
print(f'Средняя удовлетворенность работой уволившихся сотрудников: {quit.mean():4.3f}')
print(f'Средняя удовлетворенность работой оставшихся сотрудников: {stay.mean():6.3f}')
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Таким образом, статистический тест позволяет с достаточной уверенностью говорить о различии уровней удовлетворенности уволившихся и оставшихся сотрудников.

Создадим словарь для подбора параметров модели DecisionTreeClassifier (т.к. выше этот класс моделей был лучшим по целевой метрике):
- макс. глубина - от 5 до 7;
- мин. число объектов в узле - от 2 до 3;
- мин. число объектов в листе - от 2 до 4.

```python
param_grid2 = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(5, 8),
        'models__min_samples_split': range(2, 4),
        'models__min_samples_leaf': range(2, 5)
    }
]
```

Переберем различные комбинации наличия перечисленных выше категориальных признаков, подберем для них лучшую модель и рассчитаем целевую метрику на тренировочных и тестовых данных.

```python
for i in range(8):
    k = i
    ohe_columns2 = []
    if k % 2: ohe_columns2.append(ohe_columns[0])
    k = k // 2
    if k % 2: ohe_columns2.append(ohe_columns[1])
    k = k // 2
    if k % 2: ohe_columns2.append(ohe_columns[2])
    print('Используемые признаки:', ohe_columns2)
    X_train2 = X_train.drop(ohe_columns, axis=1)
    X_train2[ohe_columns2] = X_train[ohe_columns2]

    data_preprocessor2 = ColumnTransformer(
        [('ohe', ohe_pipe, ohe_columns2),
         ('ord', ord_pipe, ord_columns),
         ('num', num_pipe, num_columns)
        ],
        remainder='passthrough'
    )

    pipe_final2 = Pipeline([
        ('preprocessor', data_preprocessor2),
        ('models', DecisionTreeClassifier())
    ])

    grid_search3 = GridSearchCV(
        pipe_final2,
        param_grid2,
        cv=5,
        n_jobs=-1,
        scoring='roc_auc'
    )

    grid_search3.fit(X_train2, y_train)
    print(f'Метрика ROC-AUC для лучшей модели на тренировочных данных: {grid_search3.best_score_:.4f}')
    print()
```

Исключение признаков в данном случае приводит к ухудшению качества модели, поэтому целесообразно оставить их как есть.
