## Предварительная обработка данных

### Переименование столбцов

Переведем заголовки столбцов во всех таблицах в snake_case.

```python
def to_snakecase(text):
    text = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', text)
    text = re.sub('([a-z0-9])([A-Z])', r'\1_\2', text).lower()
    return text.replace(' ', '_')
```

```python
for name, df in dfs.items():
    df.rename(columns=to_snakecase, inplace=True)
```

### Анализ таблицы Contract

```python
contract = dfs['contract'].copy()
```

Посмотрим наличие неявных пропусков.

```python
contract['begin_date'].unique()
```

Неявные пропуски в begin_date отсутствуют.

Переведем в формат datetime.

```python
contract['begin_date'] = contract['begin_date'].astype('datetime64[ns]')
```

```python
contract['end_date'].unique()
```

Виден неявный пропуск ('No'). Заменим его на NaN и переведем в формат datetime64.

```python
contract['end_date'] = contract['end_date'] \
                       .replace('No', np.nan) \
                       .astype('datetime64[ns]')
```

```python
def is_float(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
```

```python
contract[(~contract['total_charges'].map(is_float))]
```

```python
# Функция для построения графиков числовых непрерывных параметров.
# Функция строит "сетку" графиков для различных столбцов датафрейма.
# Параметры функции:
# df - датафрейм с исходными данными;
# columns - список столбцов, по которым строятся графики;
# num_cols - число столбцов "сетке" графиков;
# title - общий заголовок для "сетки";
# hue - столбец, для уникальных значений которого строятся дополнительные графики

def plot_quant_interval(df, columns, num_cols, title, hue=None):
    num_rows = -(-len(columns) // num_cols) # округление вверх

    fig = plt.figure(figsize=(num_cols*5, num_rows*3))
    outer_grid = fig.add_gridspec(num_rows, num_cols, hspace=0.5, wspace=0.3)

    i = 0

    color = '0.8' if hue else 'C0'

    num_bins = int(1+3.222*np.log(len(df)))  # Правило Стёрджеса

    for column in columns:
        inner_grid = outer_grid[i].subgridspec(2, 1, height_ratios=[2, 1],
                                               hspace=0, wspace=0)
        ax = inner_grid.subplots(sharex=True)

        if hue:
            sns.histplot(data=df, x=column, stat='percent', common_norm=False, bins=num_bins,
                         color=color, legend=(not i), ax=ax[0], hue=hue)\
                        .set(ylabel='Частота, %')
        else:
            sns.histplot(data=df, x=column, stat='percent', common_norm=False, bins=num_bins,
                         color=color, legend=False, ax=ax[0])\
                        .set(ylabel='Частота, %')

        sns.boxplot(data=df, x=column, legend=False, ax=ax[1]).set(xlabel='Значения')

        ax[0].set_title('Признак: '+column, fontsize=12)

        # получаем параметры легенды первого графика и убираем ее
        if hue and not i:
            legend = ax[0].get_legend()
            handles = legend.legend_handles
            texts = [hue + ': ' + x.get_text() for x in legend.get_texts()]
            legend.remove()

        i += 1

    # размещаем общую легенду для сетки
    if hue:
        fig.legend(handles, texts, loc='upper center',
                   bbox_to_anchor=(0.5, -0.2), fontsize=12)

    fig.suptitle(title, va='bottom', size=16, y=1.1)
    plt.subplots_adjust(
        left=0,
        right=1,
        top=1,
        bottom=0,
        wspace=0,
        hspace=0
    )
    plt.show()

# Функция для построения графиков качественных параметров.

def plot_qual(df, columns, num_cols, title, hue=None, top=None):
    num_rows = -(-len(columns) // num_cols) # округление вверх
    fig = plt.figure(figsize=(num_cols*8, num_rows*5), layout='constrained')

    i = 0

    for column in columns:
        ax = fig.add_subplot(num_rows, num_cols, i+1)
        top_names = df[column].value_counts().iloc[:top].index
        data = df[df[column].isin(top_names)].astype('category')

        if hue:
            sns.countplot(data=data, y=column, stat='percent', legend=(not i), ax=ax,
                          hue=hue, edgecolor='black', order=top_names) \
                         .set(xlabel='Частотность, %', ylabel='Значения', title=column);
        else:
            sns.countplot(data=data, y=column, stat='percent', legend=False, ax=ax,
                          edgecolor='black', order=top_names) \
                         .set(xlabel='Частотность, %', ylabel='Значения', title=column);

        ax.set_title('Признак: '+column, fontsize=12)

        if hue and not i:
            legend = ax.get_legend()
            handles = legend.legend_handles
            texts = [hue + ': ' + x.get_text() for x in legend.get_texts()]
            legend.remove()

        i += 1

    if hue:
        fig.legend(handles, texts, loc='upper center',
                   bbox_to_anchor=(0.5, 0), fontsize=12)

    fig.set_constrained_layout_pads(wspace=0.1, hspace=0.1)
    fig.suptitle(title, va='bottom', size=16)
    plt.show()

# Функция для построения сравнительных графиков качественных параметров в зависимости от целевого признака.

def plot_qual_compare(df, columns, num_cols, title, hue=None, top=None):
    num_rows = -(-len(columns) // num_cols) # округление вверх
    fig = plt.figure(figsize=(num_cols*8, num_rows*5), layout='constrained')

    i = 0

    color = '0.8' if hue else 'C0'

    for column in columns:
        ax = fig.add_subplot(num_rows, num_cols, i+1)
        data = pd.pivot_table(df, values=hue, columns=column) \
               .T.sort_values(hue, ascending=False)[:top]
        data.index = data.index.astype(str)

        sns.barplot(data=data, orient='h', y=column, x=hue, errorbar=None, ax=ax) \
                    .set(xlabel='Доля положительного признака '+hue, ylabel='Значения', title=column)

        ax.axvline(df[hue].mean(), color='red')

        ax.set_title('Признак: '+column, fontsize=12)


        i += 1

    fig.set_constrained_layout_pads(wspace=0.1, hspace=0.1)
    fig.suptitle(title, va='bottom', size=16)
    plt.show()
```

Сформируем архитектуру нейросети с двумя скрытыми слоями и настраиваемыми параметрами.

```python
class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size_1, hidden_size_2, output_size, dropout):
        super(NeuralNet, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size_1)
        self.dp1 = nn.Dropout(p=dropout)
        self.act1 = nn.Tanh()
        self.fc2 = nn.Linear(hidden_size_1, hidden_size_2)
        self.dp2 = nn.Dropout(p=dropout)
        self.act2 = nn.Tanh()
        self.fc3 = nn.Linear(hidden_size_2, output_size)
        self.dp3 = nn.Dropout(p=dropout)
        self.act3 = nn.Sigmoid()

    def forward(self, x):
        out = self.fc1(x)
        out = self.dp1(out)
        out = self.act1(out)
        out = self.fc2(out)
        out = self.dp2(out)
        out = self.act2(out)
        out = self.fc3(out)
        out = self.dp3(out)
        out = self.act3(out)
        return out
```

Определим базовые параметры нейросети.

```python
input_size = data_preprocessor_nn.fit_transform(X_train).shape[1]
hidden_size_1 = 32
hidden_size_2 = 16
output_size = 1

do = 0.3
num_epochs = 100
learning_rate = 0.005
```

```python
net = NeuralNetBinaryClassifier(
    NeuralNet(input_size, hidden_size_1, hidden_size_2, output_size, do),
    max_epochs=num_epochs,
    criterion=nn.BCELoss(),
    optimizer=optim.Adam,
    callbacks=[EpochScoring(scoring='roc_auc', lower_is_better=False,
                            on_train=True, name='train_auc')],
    iterator_train__shuffle=True,
    verbose=0,
    train_split=False
)
```

Определим словарь параметров для перебора.

```python
param_grid = [

    # DummyClassifier
    {
        'models': [DummyClassifier(strategy='most_frequent')],
    },

    # словарь для модели LogisticRegression
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE, solver='saga')],
        'models__penalty': ['l1', 'l2'],
        'models__C': range(1, 4),
        'preprocessor': [data_preprocessor_linear] # для OHE в препроцессоре
    },

    # словарь для модели DecisionTreeClassifier
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 4),
        'models__min_samples_split': range(2, 4),
        'models__min_samples_leaf': range(1, 3)
    },

    # словарь для модели RandomForestClassifier
    {
        'models': [RandomForestClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(1, 3),
        'models__min_samples_split': range(2, 4),
        'models__min_samples_leaf': range(1, 3),
        'models__n_estimators': range(100, 200, 20)
    },

    # словарь для модели LightGBM
    {
        'models': [LGBMClassifier(random_state=RANDOM_STATE, objective='binary')],
        'models__max_depth': [2, 3],
        'models__n_estimators': [1600, 1800, 2000],
        'preprocessor': ['passthrough'] # пропускаем этап предобработки
    },

    # словарь для модели CatBoost
    {
        'models': [CatBoostClassifier(
            random_state=RANDOM_STATE,
            verbose=False,
            cat_features=cat_columns
        )],
        'models__depth': [4, 5],
        'models__iterations': [1800, 2000],
        'models__learning_rate': [0.04, 0.05],
        'preprocessor': ['passthrough'] # пропускаем этап предобработки
    },

    # словарь для нейросети
    {
        'models': [net],
        'models__lr': [0.005, 0.01],
        'preprocessor': [data_preprocessor_nn]
    }
]
```
